# Transformasi Variabel

```python
#List kolom variabel kategorikal
hn.columns
```

Transformasi pada variabel SalePrice dan LotArea

```python
spt = np.log10(hn['SalePrice'])
sns.distplot(spt, axlabel=spt.name)
line1 = plt.axvline(spt.mean(), color='k', linestyle='--', label="mean")
line2 = plt.axvline(spt.median(), color='r', linestyle='--', label="median")
plt.legend(handles=[line1, line2], loc=1)  
print("Median:", spt.median())
print("Mean:", spt.mean())
print("Std Dev:", spt.std())
plt.title('Plot Distribusi Transformasi SalePrice')
plt.show()
```

Berdasarkan output yang diberikan:

1. Data yang ditransformasi secara logaritmik cenderung memiliki distribusi yang lebih simetris, dengan mean dan median yang relatif dekat.
2. Standar deviasi yang rendah menunjukkan bahwa data tidak terlalu tersebar di sekitar mean.
3. Skew yang sedikit ke kanan (mean lebih besar dari median) mungkin menunjukkan bahwa meskipun transformasi telah membantu, masih ada sedikit skewing (Skew ke kanan).

```python
lpt = np.log10(hn['LotArea'])
sns.distplot(lpt, axlabel=lpt.name)
line1 = plt.axvline(lpt.mean(), color='k', linestyle='--', label="mean")
line2 = plt.axvline(lpt.median(), color='r', linestyle='--', label="median")
plt.legend(handles=[line1, line2], loc=1)  
print("Median:", lpt.median())
print("Mean:", lpt.mean())
print("Std Dev:", lpt.std())
plt.title('Plot Distribusi Transformasi Lot Area')
plt.show()
```

Berdasarkan output yang diberikan:

1. Data yang ditransformasi secara logaritmik cenderung memiliki distribusi yang lebih simetris, dengan mean dan median yang relatif dekat.
2. Standar deviasi yang rendah menunjukkan bahwa data tidak terlalu tersebar di sekitar mean.
3. Skew yang sedikit ke kiri (mean lebih kecil dari median) mungkin menunjukkan bahwa meskipun transformasi telah membantu, masih ada sedikit skewing (Skew ke kiri).

# 2D Visualization

```python
#Korelasi SalePrice dengan GrLivArea
grs=sns.scatterplot(y='SalePrice',x='GrLivArea',data=hn)
plt.title('Korelasi SalePrice dan GrlivArea')
```

Dari output di atas, dapat dilihat bahwa variabel 'SalePrice' memiliki korelasi positif dengan variabel GrLivArea dikarenakan plot plot data membentuk garis yang cenderung ke atas, yang artinya semakin meningkat nilai SalePrice, maka semakin meningkat pula GrLivArea. 

Meskipun pada grafik scatterplot kedua variabel tersebut berkorelasi, kita tetap perlu melakukan pengujian formal untuk menghitung nilai korelasi dengan tujuan mendapatkan hasil yang lebih pasti, karena setiap orang memiliki pemahaman yang berbeda pada grafik.

```python
corr_SaleGrLiv = hn['SalePrice'].corr(hn['GrLivArea'])
corr_SaleGrLiv
```

Berdasarkan output, nilai korelasi antara SalePrice dengan GrlivArea adalah 0.708 yang artinya kedua variabel tersebut memiliki korelasi positif yang kuat. Hubungan antara variabel sangat jelas, dan perubahan pada satu variabel biasanya disertai perubahan yang cukup signifikan pada yang lainnya.

Berdasarkan output, nilai korelasi antara SalePrice dengan GarageArea adalah 0.608 yang artinya kedua variabel tersebut memiliki korelasi positif yang kuat. Hubungan antara variabel sangat jelas, dan perubahan pada satu variabel biasanya disertai perubahan yang cukup signifikan pada yang lainnya.

Note: Rentang Nilai Korelasi
1. 0 hingga 0.2: Korelasi sangat lemah. Hubungan antara variabel sangat rendah, dan perubahan pada satu variabel memiliki pengaruh minimal pada yang lainnya.
2. 0.2 hingga 0.4: Korelasi lemah. Ada hubungan yang bisa diamati, tetapi pengaruhnya tidak terlalu signifikan.
3. 0.4 hingga 0.6: Korelasi sedang. Ada hubungan yang lebih jelas antara variabel, dan perubahan pada satu variabel dapat mempengaruhi yang lainnya.
4. 0.6 hingga 0.8: Korelasi kuat. Hubungan antara variabel sangat jelas, dan perubahan pada satu variabel biasanya disertai perubahan yang cukup signifikan pada yang lainnya.
5. 0.8 hingga 1: Korelasi sangat kuat. Hubungan antara variabel hampir sempurna, dan perubahan pada satu variabel hampir selalu disertai perubahan pada variabel lainnya dengan pola linear yang kuat.
